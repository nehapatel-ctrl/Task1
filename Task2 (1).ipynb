```python
# Question 2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/Titanic-Dataset.csv')  # Change the path if needed

# Select numeric columns
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

# --- Histograms ---
df[numeric_cols].hist(bins=20, figsize=(12, 8), edgecolor='black')
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()

# --- Boxplots ---
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numeric_cols])
plt.title('Boxplots of Numeric Features')
plt.xlabel('Features')
plt.tight_layout()
plt.show()
```

```python
# Question 3
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix for numerical columns
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
# Pairplot with 'Survived' as the hue
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', palette='husl')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
# Question 4
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.kdeplot(df[df['Survived'] == 1]['Age'], label='Survived', fill=True)
sns.kdeplot(df[df['Survived'] == 0]['Age'], label='Not Survived', fill=True)
plt.title('Age Distribution by Survival')
plt.legend()
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")
plt.show()
```
